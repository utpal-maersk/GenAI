```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import joblib
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,f1_score,confusion_matrix,recall_score

data=pd.read_csv('student_train_logistic.csv')
data.head()

#convert the text column city college into numberic

data=pd.get_dummies(data,columns=["City","College"])
data.head(25)
```

```python
#define x,y
data=data.drop("Student_ID",axis=1)
X = data.drop("Placed",axis=1)
Y = data[['Placed']]
data.head()

#split the data into training and test
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)

```

```python
#train the model
model=LogisticRegression()
model.fit(X_train,Y_train)
joblib.dump(model, 'student_placement.pkl')
```

```python
#predict on test data

y_predict=model.predict(X_test)
print(y_predict)
print(Y_test)
print("Accuracy: ",accuracy_score(Y_test,y_predict))
print("Pressicion",precision_score(Y_test,y_predict))
print("Recall",recall_score(Y_test,y_predict))
print("F1 ",f1_score(Y_test,y_predict))
print("Confusion ",confusion_matrix(Y_test,y_predict))
```

```python
#predict on new data

encoded_values=pd.read_csv("student_new_logistic.csv")
encoded_values.head()
encoded_values=pd.get_dummies(encoded_values,columns=["City","College"])
encoded_values.head()
encoded_values=encoded_values.drop("Student_ID",axis=1)
encoded_values.head()
encoded_values["Predicted_placement"]=model.predict(encoded_values)
encoded_values.head()
encoded_values["Predicted_placement"]=encoded_values["Predicted_placement"].map({0:"Not Placed",1:"Placed"})
encoded_values.head(20)
encoded_values.to_csv("Predicted_student_placement.csv", index=False)
print(" Predictions saved to Predicted_student_placement.csv successfully!")
encoded_values.head()
```
