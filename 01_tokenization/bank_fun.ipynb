```python
customers={}

def addCustomer(AccountNo,Name,Age,MobileNo):
    if AccountNo in customers:
        print("Customer already exist")
    else:
        customers[AccountNo]={"Name":Name,"Age" :Age,"MobileNo":MobileNo}
        print("Customer Added Successfully !!")

def displayCustomerDetails(AccountNo):
     if AccountNo in customers:
        print(f"Customer Account No is {AccountNo}")
        print(f"Customer Name  is {customers[AccountNo]['Name']}")
        print(f"Customer Age  is {customers[AccountNo]['Age']}")
        print(f"Customer Mobile No is {customers[AccountNo]['MobileNo']}")
     else:
        print(f"Customer Account No is not Exist")


addCustomer(53600,'Utpal',45,9940909304)
displayCustomerDetails(53601)


```
